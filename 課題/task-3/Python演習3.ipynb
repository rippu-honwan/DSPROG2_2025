### 問1

自然数を入力とし、入力された自然数が3の倍数(Ex. 3, 6, 9, ...)であるときは`hoge`、3を含む数(Ex. 3, 13, 23, ...)であるときは`huga`、両方の条件を満たす時は`piyo`を出力する関数`nabeatsu`を定義してください

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
6

# テストケース2
13

# テストケース3
33
```

````python
def nabeatsu(num: int) -> str:
    # 3の倍数かどうか判定
    is_multiple_of_3 = (num % 3 == 0)
    # 3を含む数字かどうか判定
    has_3 = ('3' in str(num))
    
    # 両方の条件を満たす場合
    if is_multiple_of_3 and has_3:
        return 'piyo'
    # 3の倍数である場合
    elif is_multiple_of_3:
        return 'hoge'
    # 3を含む数字である場合
    elif has_3:
        return 'huga'
    # どちらの条件も満たさない場合（問題文には明記されていませんが、通常は数字をそのまま返します）
    else:
        return str(num)

# テストケースの実行
print('case_1:', nabeatsu(6))   # 3の倍数 -> hoge
print('case_2:', nabeatsu(13))  # 3を含む -> huga
print('case_3:', nabeatsu(33))  # 両方 -> piyo
````

````python
case_1 = 6
case_2 = 13
case_3 = 33

print('case_1:', nabeatsu(case_1))
print('case_2:', nabeatsu(case_2))
print('case_3:', nabeatsu(case_3))
````

### 問2

異なる自然数を含む配列とターゲットとなる自然数を入力として、配列に含まれる自然数を足し合わせてターゲットとなる自然数になる組み合わせを出力する関数`combination_sum`を定義してください

このとき、同じ自然数は何度でも利用してよいものとし、もし組み合わせが存在しない場合は空の配列を返してください

例えば、配列`[2, 3, 6, 7]`、ターゲット`7`を入力としたときの出力は`[[2,2,3],[7]]`になります

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
配列: [2, 3, 6, 7]
ターゲット: 14

# テストケース2
配列: [2, 3, 5]
ターゲット: 8

# テストケース3
配列: [2, 9, 11]
ターゲット: 100
```

````python
def combination_sum(candidates: list, target: int) -> list:
    results = []
    
    # 探索を効率化するためにソートする
    candidates.sort()
    
    def backtrack(remaining, start_index, path):
        # 目標値に達した場合、結果リストに追加
        if remaining == 0:
            results.append(list(path))
            return
        
        for i in range(start_index, len(candidates)):
            num = candidates[i]
            
            # 残りの値より現在の数字が大きい場合、それ以降の探索は不要（ソート済みのため）
            if num > remaining:
                break
            
            # 数字を選択して再帰呼び出し
            path.append(num)
            # 同じ数字を何度でも使えるため、インデックスは i のまま
            backtrack(remaining - num, i, path)
            # バックトラック（戻る）
            path.pop()
            
    backtrack(target, 0, [])
    return results

# テストケースの実行
candidates_1 = [2, 3, 6, 7]
target_1 = 14
print(combination_sum(candidates_1, target_1))

candidates_2 = [2, 3, 5]
target_2 = 8
print(combination_sum(candidates_2, target_2))

candidates_3 = [2, 9, 11]
target_3 = 100
# 結果が長くなるため、ここでは個数のみ確認（必要に応じてprintしてください）
print(f"Combinations found for case 3: {len(combination_sum(candidates_3, target_3))}")
````

````python
candidates_1 = [2, 3, 6, 7]
target_1 = 14

candidates_2 = [2, 3, 5]
target_2 = 8

candidates_3 = [2, 9, 11]
target_3 = 100

print(combination_sum(candidates_1, target_1))
print(combination_sum(candidates_2, target_2))
print(combination_sum(candidates_3, target_3))
````

### 問3

括弧を含む文字列を入力として、括弧の内包関係が正しいかどうかを判定し、正しければ`True`、正しくなければ`False`を返す関数`format_checker`を定義してください

例えば、`[](){}`や`[({})]`は正しい、`[(]`や`{[(})]`は正しくないと判定します

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
Hello I’m [(firstname) (lastname)] and I’m {age} years old!

# テストケース2
Hello I’m [(firstname) (lastname]) and I’m {age} years old!

# テストケース3
Hello I’m ((firstname) (lastname) and I’m {age} years old!
```

````python
def format_checker(text: str) -> bool:
    stack = []
    # 閉じ括弧と開き括弧の対応マップ
    mapping = {")": "(", "]": "[", "}": "{"}
    
    for char in text:
        # 開き括弧の場合、スタックに追加
        if char in mapping.values():
            stack.append(char)
        # 閉じ括弧の場合
        elif char in mapping.keys():
            # スタックが空、または対応する開き括弧がスタックのトップにない場合はFalse
            if not stack or mapping[char] != stack.pop():
                return False
    
    # 処理終了後、スタックが空であれば全ての括弧が正しく閉じられている
    return not stack

# テストケースの実行
case_1 = 'Hello I’m [(firstname) (lastname)] and I’m {age} years old!'
case_2 = 'Hello I’m [(firstname) (lastname]) and I’m {age} years old!'
case_3 = 'Hello I’m ((firstname) (lastname) and I’m {age} years old!'

print('case_1:', format_checker(case_1))
print('case_2:', format_checker(case_2))
print('case_3:', format_checker(case_3))
````
